```python
import numpy as np
import statistics
from scipy.stats import norm, t
```

```python
def confidence_interval(data, alpha, method):
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)

    if method == "z":
        crit = norm.ppf(1 - alpha/2)
        half = crit * std / np.sqrt(n)
    else:  # "t"
        crit = t.ppf(1 - alpha/2, df=n-1)
        half = crit * std / np.sqrt(n)

    return mean - half, mean + half, crit
```

```python
data = input("data (comma-separated): ")
data = np.array([float(x) for x in data.split(",")])
alpha = float(input("alpha: "))
method = input("method (z or t): ")

avg = sum(data) / len(data)
sample_stddev = statistics.stdev(data)

print("---------------------------------------")

ci_low, ci_high, crit = confidence_interval(data, alpha, method)

print("critical value:", crit)
print("Confidence Interval:", (ci_low, ci_high))

print("mean:", avg)
print("sample s:", sample_stddev)
```

```python
import numpy as np
from scipy import stats
from scipy.stats import t
import math

def willink_ci_law(data, alpha):
    X = np.asarray(data, dtype=float)
    n = len(X)
    
    sample_mean = np.mean(X)
    sample_var = np.var(X, ddof=1)
    
    # third moment
    sum_cubed_deviations = np.sum((X - sample_mean)**3)
    mu3_hat = (n / ((n - 1) * (n - 2))) * sum_cubed_deviations
    
    # Willink skewness parameter
    a = mu3_hat / (6 * np.sqrt(n) * (sample_var ** 1.5))
    
    # t quantile
    tval = t.ppf(1 - alpha/2, df=n-1)
    
    # G-function
    def G(r, a):
        return (np.cbrt(1 + 6*a*(r - a)) - 1) / (2*a)
    
    se = np.sqrt(sample_var / n)
    
    lower = sample_mean - G(tval, a) * se
    upper = sample_mean - G(-tval, a) * se

    half_l = sample_mean - lower
    half_u = upper - sample_mean
    
    return lower, upper, half_l, half_u

Y = df["Y"]

lower, upper, half_l, half_u = willink_ci_law(Y, 0.01)

print("Willink Interval meu_Y:", lower,",", upper)
print("lower half length:", half_l)
print("upper half length:", half_u)
```

```python
import numpy as np
from scipy.stats import norm
from scipy.stats import skew

n = 100000

# Generate x,y
X = np.random.uniform(0, 1, n)
Y = np.random.uniform(0, 1, n)

# Generate B: points within both of two circles
# ChatGPT 5.2: asked it to generate numbers 0 or 1 
# if a number x, y is within the following circles 
# (centered on 0,0 with r = 1), (centered on 1,1 with r = 1) in python 

B = ((X**2 + Y**2 <= 1) &
     ((X - 1)**2 + (Y - 1)**2 <= 1)).astype(int)


# A_d
A_d = np.mean(B)

alpha = 0.05
z = norm.ppf(1 - alpha/2)
half_length = z * np.sqrt(A_d * (1 - A_d) / n)

# confidence interval
ci_lower = A_d - half_length
ci_upper = A_d + half_length

skew_B = skew(B)

# Print
print(f"Estimated A_D: {A_d:.6f}")
print(f"z-Confidence Interval, alpha = 0.05, for A_D: ({ci_lower:.6f}, {ci_upper:.6f})")
print(f"Half-length: {half_length:.6f}")
print(f"Estimated skewness of B: {skew_B:.6f}")
```
